```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from census import Census

```

```python
# Census API Key
from configc import census_key
c = Census(census_key, year=2020)
```

```python
# Run Census Search to retrieve data on all zip codes (2020 ACS5 Census)
# See: https://github.com/CommerceDataService/census-wrapper for library documentation
# See: https://gist.github.com/afhaque/60558290d6efd892351c4b64e5c01e9b for labels
census_data = c.acs5.get(("NAME", "B25097_001E", "B25096_001E", "B25099_001E",
                          "B25035_001E",
                          "B25081_001E"), {'for':'state:06,32'})

# Convert to DataFrame
census_pd = pd.DataFrame(census_data)

# Column Reordering
census_pd = census_pd.rename(columns={"B25097_001E": "Mortgage Status By Median Value",
                                      "B25096_001E": "Mortgage Status By Value",
                                      "B25035_001E": "Median Year Structure Built",
                                      "B25099_001E": "Mortgage Status By Median Household Income",
                                      "B25081_001E": "Mortgage Status",
                                      "NAME": "Name", "state": "State"})



# Final DataFrame
census_pd = census_pd[["State", "Mortgage Status", "Mortgage Status By Median Value", "Mortgage Status By Median Household Income",
                        "Median Year Structure Built"]]

# Visualize
print(len(census_pd))
census_pd.head()
```

```python
# Create an array that contains the number of mortgage status median value that California had
med_value = [479200, 458500, 383900, 385500, 443400, 538500]
x_axis = np.arange(len(med_value))
years=["2009", "2010", "2012", "2015", "2017","2020"]
```

```python
# Tell matplotlib that we will be making a bar chart
# Med_value is our y axis and x_axis is, of course, our x axis
# We apply align="edge" to ensure our bars line up with our tick marks
plt.bar(x_axis, med_value, color='r', alpha=0.5, align="center")
```

```python
# Tell matplotlib where we would like to place each of our x axis headers
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, years)
```

```python
# Sets the x limits of the current chart
plt.xlim(-0.75, len(x_axis)-0.25)
```

```python
# Sets the y limits of the current chart
plt.ylim(0, max(med_value)+15000)
```

```python
# Give our chart some labels and a tile
plt.title("CA Mortgage Status")
plt.xlabel("Years")
plt.ylabel("Mortgage Status by Median Value")
```
